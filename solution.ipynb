Task 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

df = pd.read_csv('Salaries.csv')
numRows = len(df.axes[0])
numcols = len(df.axes[1])

print("Number of Rows: ",numRows)
print("Number of columns: ",numcols, '\n')
print(df.dtypes,'\n')
print(df.isnull())
```

Task 2

```python
#claculations isn't accurate because of the missing and null values 
slryMean = df[df.columns[7]].mean()
slryMedian = df[df.columns[7]].median()
slryMode = df[df.columns[7]].mode()
slryMin = df[df.columns[7]].min()
slryMax = df[df.columns[7]].max()
slryRange = slryMax - slryMin
slryStd = df[df.columns[7]].std()


print("The mean of the salary: ",slryMean)
print("The median of the salary: ",slryMedian)
print("The mode of the salary: ",slryMode)
print("The minimum Number of the salary: ",slryMin)
print("The maximum Number of the salary: ",slryMax)
print("The Range of the salary: ",slryRange)
print("The standard deviation of the salary: ",slryStd)
```

Task 3

```python
# print(df['BasePay'].isnull().sum()) #609
# print(df['OvertimePay'].isnull().sum()) #4
# print(df['OtherPay'].isnull().sum())  #4
# print(df['Notes'].isnull().sum()) #36163

baseMean = df[df.columns[3]].mean()
overTPMean = df[df.columns[4]].mean()
otherMean = df[df.columns[5]].mean()
BenfMean = df[df.columns[6]].mean()

df["BasePay"].fillna(baseMean, inplace= True)
df["OvertimePay"].fillna(overTPMean, inplace= True)
df["OtherPay"].fillna(otherMean, inplace= True)
df["Benefits"].fillna(BenfMean, inplace=True)


#replacing the missing data with the mean for it's column as it's one of the simple solution for data cleaning
#for columns "Notes" and "Status" we can drop the columns or leave them as is
```

Task 4

```python
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20)
plt.title('Salaries')
plt.xlabel('Total Pay')
plt.show()

deptCount = df['Agency'].value_counts()
plt.pie(deptCount, labels=deptCount.index, autopct='%1.1f%%')
plt.show()
```

Task 5

```python
# print(df.groupby(["JobTitle", "Year"])["BasePay"].describe())

average = df.groupby(["Year"])["BasePay"].mean()
average.plot(kind='bar')
plt.ylabel('Average Base Pay')
plt.xlabel('Years')
plt.show()
```

Task 6

```python
correlation = df['TotalPay'].corr(df['BasePay'])

print("The correlation between TotalPay and BasePay: ", correlation)

plt.scatter(df['BasePay'].index, df['BasePay'])
plt.scatter(df['TotalPay'].index, df['TotalPay'])
plt.title('TotalPay vs BasePay')

plt.show()
```
